```python
# Description of the data
# The given dataset has 1309 rows and 28 columns, of which most of them are zero columns that I have no idea how to use them
# I therefore internd to drop those columns in data cleaning to easily manage my data into what I want to use it for.
# Also realized that the column for sex is also in binary, taking 0 for males and 1 for females, also want to reverse that to visualize my data well.
```

```python
# Data cleaning

# Drop the unwanted columns
new_df = titanic_df.drop(titanic_df.columns[[5, 6, 7, 8, 9, 10, 11, 13, 14, 15, 16, 17, 18, 19, 20, 22, 23, 25, 26]], axis=1)
new_df.head()
```

```python
# Renaming a misspelled column
renamed_df = new_df.rename(columns={'2urvived': 'Survived'})
renamed_df.head()
```

```python
# Adding the names of the passengers
names_df = pd.read_csv('name.csv')
names_df.head()
```

```python
# Add the names to the dataset
merged_df = pd.merge(renamed_df, names_df, on='Passengerid', how='left')
merged_df.head()
                
```

```python
# Re-order the columns
# Get the current column order
columns = merged_df.columns.tolist()

# Temporarily remove the name column
columns.remove('Name')

#Insert the desired order
columns.insert(columns.index('Passengerid') + 1, 'Name')

# Re-order the DataFrame
merged_df = merged_df = merged_df[columns]
merged_df.head()
```

```python
# Save the cleaned dataset
merged_df.to_csv('cleaned_dataset1.csv', index=False)
```
